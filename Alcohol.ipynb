# Alcohol Consumption, Happiness, and Economic Factors

## Problem Statement
This project investigates the relationship between alcohol consumption, happiness levels, and GDP per capita across countries.

## Research Questions
- Is alcohol consumption correlated with happiness?
- Is there a relationship between alcohol consumption, GDP per capita, and happiness across countries?
- Can we predict happiness using economic and behavioral factors?

## Data Sources
- Alcohol consumption dataset
- World Happiness Report
- GDP per capita (World Bank)

## Goal
To analyze and model the relationship between these variables using statistical and machine learning methods.
